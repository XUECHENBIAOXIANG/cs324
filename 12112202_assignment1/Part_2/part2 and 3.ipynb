```python
import train_mlp_numpy as MLP
import time
DNN_HIDDEN_UNITS_DEFAULT = '20'
LEARNING_RATE_DEFAULT = 1e-2
MAX_EPOCHS_DEFAULT = 1500
EVAL_FREQ_DEFAULT = 10
num_samples = 1000
num_features = 10
num_classes = 2
begin = time.time()
```

```python
X, y,accuracy_list,accuracy_list_test,cost_step,loss_list,loss_list_test =MLP.train(DNN_HIDDEN_UNITS_DEFAULT,LEARNING_RATE_DEFAULT,MAX_EPOCHS_DEFAULT,EVAL_FREQ_DEFAULT )
print( time.time()-begin)
print(cost_step)
```

```python
import matplotlib.pyplot as plt

# 
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
plt.title('Scatter Plot of Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1500), accuracy_list, label='Training Accuracy')
plt.plot(range(1500), accuracy_list_test, label='Test Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.plot(range(1500),loss_list, label='Training loss')
plt.plot(range(1500), loss_list_test, label='Test loss')
plt.title('loss Curves')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend()
plt.grid(True)
plt.show()

```
